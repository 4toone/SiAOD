### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = input()
user_n = input()
user_min_limit = input()
user_max_limit = input()
user_m, user_n, user_min_limit, user_max_limit

def matritsa(m, n, min_limit, max_limit):
    result_matritsa = []
    if int(m) > 0:
        for i in range(int(m)):
            polu_matritsa = []
            if int(n) > 0:
                for j in range(int(n)):
                    polu_matritsa.append(random.randint(int(min_limit), int(max_limit)))
            result_matritsa.append(polu_matritsa) 
    return result_matritsa

gotovaya_matritsa = matritsa(user_m, user_n, user_min_limit, user_max_limit)

def korrektirovka_matritsi(matrica):
    for row in matrica:
        for element in row:
            print("\t" + str(element), end = " ")
        print()     


korrektirovka_matritsi(gotovaya_matritsa)

            
    
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO
matrix = copy.deepcopy(gotovaya_matritsa)
def selection_sort(stroka):
    i = 0
    while i < len(stroka) - 1:
        mindex = i
        j = i + 1
        while j < len(stroka):
            if stroka[j] < stroka[mindex]:
                mindex = j
            j += 1
        stroka[i], stroka[mindex] = stroka[mindex], stroka[i]
        i += 1

t1 = time.time()        
        
for strokka in matrix:
    strokka = selection_sort(strokka)

t2 = time.time()
selection_sort_time = round((t2 - t1) * 1000)     

korrektirovka_matritsi(matrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
#TODO
matrix = copy.deepcopy(gotovaya_matritsa)
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

t1 = time.time()
        
for strokka in matrix:
    strokka = insertion_sort(strokka)
    
t2 = time.time()
insertion_sort_time = round((t2 - t1) * 1000)     

korrektirovka_matritsi(matrix)        
        
        
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
#TODO
matrix = copy.deepcopy(gotovaya_matritsa)
def bubble_sort(arr):
    n = 1
    while n < len(arr):
        for k in range(len(arr) - n):
            if arr[k] > arr[k + 1]:
                arr[k], arr[k + 1] = arr[k + 1], arr[k]
        n = n + 1

t1 = time.time()        
        
for strokka in matrix:
    strokka = bubble_sort(strokka)

t2 = time.time()
bubble_sort_time = round((t2 - t1) * 1000)      
    
korrektirovka_matritsi(matrix)        
    

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
#TODO
matrix = copy.deepcopy(gotovaya_matritsa)
def shell_sort(arr):
    gap = len(arr)//2
    while gap > 0: 
        for i in range(gap, len(arr)):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:    
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2
        
t1 = time.time()  

for strokka in matrix:
    strokka = shell_sort(strokka)

t2 = time.time()
shell_sort_time = round((t2 - t1) * 1000)  

korrektirovka_matritsi(matrix)  
            

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
#TODO
matrix = copy.deepcopy(gotovaya_matritsa)
def quick_sort(strok):
    if len(strok) >= 1:
        return strok
    pivot = strok[-1]
    wall = 0
    for current in range(len(strok) - 1):
        if strok[current] < pivot:
            strok[wall], strok[current] = strok[current], strok[wall]
            wall = wall + 1
    strok.insert(wall, pivot)
    strok.pop(-1)
    return quick_sort(strok[0: wall]) + [strok[wall]] + quick_sort(strok[(wall + 1): len(strok)])

t1 = time.time()

new_matrix = []    
for stroka in matrix:
    new_matrix.append(tournament_sort(stroka))

t2 = time.time()
quick_sort_time = round((t2 - t1) * 1000)    
    
korrektirovka_matritsi(new_matrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
#TODO

import heapq
matrix = copy.deepcopy(gotovaya_matritsa)

def tournament_sort(strok):
    heap = []
    heapsize = 10
    if len(strok) >= heapsize:
        heap = strok[:heapsize]
    else:
        heap = strok[:]
    heapq.heapify(heap)
    
    winners = [float('-inf')]
    losers = []
    
    i = heapsize
    while i < len(strok) or len(heap) > 0:
        heap_top = heapq.heappop(heap)
        if heap_top >= winners[-1]:
            winners.append(heap_top)
        else:
            losers.append(heap_top)
        
        if i < len(strok):
            heapq.heappush(heap, strok[i])
        i += 1
    
    del winners[0]
    strok = losers + winners

    if len(losers) > 0:
        return tournament_sort(strok)
    return strok

t1 = time.time()

new_matrix = []    
for stroka in matrix:
    new_matrix.append(tournament_sort(stroka))

t2 = time.time()
tournament_sort_time = round((t2 - t1) * 1000)       
    
korrektirovka_matritsi(new_matrix)  

    
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
